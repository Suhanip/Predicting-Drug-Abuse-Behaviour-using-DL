```python
import pandas as pd
```

```python
for flag in range(1,405):    
    file = open("ldatimeline/tweets"+str(flag)+".txt","r")
    lines = []
    for f in file:
        lines.append(f)
    file.close()
    txt = ""
    for i in lines[0]:
        if i not in ['[',']','(',')','\n']:
            txt += i
    lines[0] = txt
    lines[0] = lines[0].split(",")
    words = []
    for i in range(1,len(lines)):
        l = lines[i].split(" ")
        l.remove(l[0])
        l[len(l)-1] = l[len(l)-1][:-1]
        words.append(l)
    file = open("lda2bert/tweets"+str(flag)+".txt","a")
    for line in words:
        line = " ".join(line)
        file.write(line+'\n')
    file.close()
```
